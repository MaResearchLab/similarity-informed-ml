```python
%%time

from scatter_fit import PeakFitter
import numpy as np

if __name__ == "__main__":
    fitter = PeakFitter(
        sample="new_chems_full", # Name of the excel sheet with 1D data
        data_dir="raw_1d/", # Directory of the 1D data
        output_dir="outputs/", # Output folder for fitted data
        window_range=np.linspace(0.004, 0.01, 20), # Set the window fitting range (Should be set to larger than the peak width)
        alpha=10, # Larger values for alpha incentivizes larger fitting windows
        low_q_limit=0.002,  # Set the value to trim the noisy low q region of the data
        q_and_I_in_pairs=True # True if there is a q for every I, otherwise False for only one q for each I
    )
    fitter.fit_all()
```
