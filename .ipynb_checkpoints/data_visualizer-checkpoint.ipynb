# Import packages and load data

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib import cm
from matplotlib import rc
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)  
from sklearn.preprocessing import StandardScaler
from matplotlib.colors import Normalize
from sklearn.manifold import TSNE
```

```python
directory_path = './chemical-featurization/likeness_data/'
output_dir = './figures/'
similarity_methods = ["morgan", "rdkit_path", "maccs", "rdkit_physchem", "mordred"]
smiles_maps = ["poly", "repeat", "monomer"]
df_likeness = {}
for name in smiles_maps:
    for method in similarity_methods:
        # Continuous descriptor methods cannot handle the wildcard present in the 'poly' SMILES
        if "poly" in name.lower() and method in ["rdkit_physchem", "mordred"]:
            continue
         # Set the key for each dataset
        df_likeness_key = f"{name}_{method}"
        file_path = f"{directory_path}{df_likeness_key}.xlsx"
        df_likeness[df_likeness_key] = pd.read_excel(file_path)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from matplotlib.ticker import AutoMinorLocator
from matplotlib import rc

# --------------------------------------------------
# Classification
# --------------------------------------------------
def classify(df):
    min_AA, max_AA = df['comp1_blockA_type_comp2_blockA_type_exists'].min(), df['comp1_blockA_type_comp2_blockA_type_exists'].max()
    min_AB, max_AB = df['comp1_blockA_type_comp2_blockB_type_exists'].min(), df['comp1_blockA_type_comp2_blockB_type_exists'].max()
    min_BB, max_BB = df['comp2_blockA_type_comp2_blockB_type_exists'].min(), df['comp2_blockA_type_comp2_blockB_type_exists'].max()

    cond_AA = df['comp1_blockA_type_comp2_blockA_type_exists']
    cond_AB = df['comp1_blockA_type_comp2_blockB_type_exists']
    cond_BB = df['comp2_blockA_type_comp2_blockB_type_exists']

    types = np.full(len(df), 'BCP+BCP', dtype=object)
    types[(cond_AA == max_AA) & (cond_BB == min_BB)] = 'BCP+HP'
    types[(cond_AB == max_AB)] = 'BCP'
    return types


# --------------------------------------------------
# Data
# --------------------------------------------------
y1 = targets['gisaxs_fwhm']       # X-axis of joint plot
y2 = targets['gisaxs_domain']     # Y-axis of joint plot
classes = classify(inputs)
```

```python
# Plot style
%matplotlib inline
rc('text', usetex=False)
rc('mathtext', fontset='cm')
rc('xtick', labelsize=25)   
rc('xtick.major', size=7)  
rc('xtick.minor', size=4)
rc('xtick.major', width=2)
rc('xtick.minor', width=2)
rc('ytick.major', width=2)
rc('ytick.minor', width=2)
rc('ytick', labelsize=25)
rc('ytick.major', size=7)
rc('ytick.minor', size=4)
rc('axes', labelsize=25) 
rc('axes', linewidth=2) 
rc('font',family='sans serif')
rc('font', style='normal')
rc('font', weight='500')
rc('font', size='15')
rc('axes', labelweight='500')
rc('axes.spines', **{'right':True, 'top':True}) 
plt.rcParams['font.family'] = 'Arial'

# Figure layout
fig = plt.figure(figsize=(16, 8), constrained_layout=True)
gs = fig.add_gridspec(
    2, 2,
    width_ratios=(7, 1.5),
    height_ratios=(1.5, 7),
    hspace=0.0,
    wspace=-0.005
)

ax_joint = fig.add_subplot(gs[1, 0])
ax_top   = fig.add_subplot(gs[0, 0], sharex=ax_joint)
ax_right = fig.add_subplot(gs[1, 1], sharey=ax_joint)

# Class colors
colors = {'BCP': 'indigo', 'BCP+HP': 'dodgerblue', 'BCP+BCP': 'forestgreen'}

# Joint scatter
for label, color in colors.items():
    mask = classes == label
    ax_joint.scatter(
        y1[mask],
        y2[mask],
        s=200,
        color=color,
        edgecolors='black',
        alpha=0.75,
        label=label
    )

ax_joint.set_xlabel("FWHM ($A^{-1}$)", fontsize=35)
ax_joint.set_ylabel("Domain spacing (nm)", fontsize=35)
ax_joint.legend(frameon=False, fontsize=20)

# Density along X (FWHM) — FILLED
x_lims = (y1.min(), y1.max())
x_grid = np.linspace(x_lims[0], x_lims[1], 200)

for label, color in colors.items():
    vals = y1[classes == label]
    if len(vals) > 1:
        kde = gaussian_kde(vals)
        dens = kde(x_grid)
        ax_top.plot(x_grid, dens, color=color, lw=1, alpha=0.95)
        ax_top.fill_between(x_grid, 0, dens, color=color, alpha=0.5)

ax_top.axis('off')

# Density along Y (Domain spacing) — FILLED
y_lims = (y2.min(), y2.max())
y_grid = np.linspace(y_lims[0], y_lims[1], 200)

for label, color in colors.items():
    vals = y2[classes == label]
    if len(vals) > 1:
        kde = gaussian_kde(vals)
        dens = kde(y_grid)
        ax_right.plot(dens, y_grid, color=color, lw=1, alpha=0.95)
        ax_right.fill_betweenx(y_grid, 0, dens, color=color, alpha=0.5)

ax_right.axis('off')

# Final aesthetics
ax_joint.xaxis.set_major_locator(MultipleLocator(.001))
ax_joint.xaxis.set_minor_locator(AutoMinorLocator(2))
ax_joint.yaxis.set_major_locator(MultipleLocator(10))
ax_joint.yaxis.set_minor_locator(AutoMinorLocator(2))

for spine in ax_joint.spines.values():
    spine.set_color('black')
ax_joint.grid(False)
ax_joint.set_facecolor("white")
ax_joint.tick_params(
    axis='both',
    which='both',
    direction='out',
    bottom=True,
    left=True,
    length=7,
    width=2
)
# plt.tight_layout()
plt.savefig(output_dir + fig_title, format='png', dpi=300, transparent=True)
plt.show()
```

# Composite blend effects

```python
x1 = inputs['comp1_frac']
x2 = inputs['add_ratio']
y1 = targets['gisaxs_fwhm']
y2 = targets['gisaxs_domain']

bc34_x1 = x1[320:344]
bc34_x2 = x2[320:344]
bc34_y1 = y1[320:344]
bc34_y2 = y2[320:344]

hp_x1 = x1[-78:]
hp_x2 = x2[-78:]
hp_y1 = y1[-78:]
hp_y2 = y2[-78:]

full_hp_x1 = np.concatenate([bc34_x1, hp_x1])
full_hp_x2 = np.concatenate([bc34_x2, hp_x2])
full_hp_y1 = np.concatenate([bc34_y1, hp_y1])
full_hp_y2 = np.concatenate([bc34_y2, hp_y2])

# Filter for additive ratio > 4
mask = full_hp_x2 > 2
# mask = (full_hp_x2 > 4) & (full_hp_x2 < 5)


filtered_x1_hp = full_hp_x1[mask]
filtered_x2_hp = full_hp_x2[mask]
filtered_y2_hp = full_hp_y2[mask]
filtered_y1_hp = full_hp_y1[mask]

# print(f'Length of bc: {len(bc_y2)}, length of hp: {len(hp_y2)}, length of both: {len(combo_x1)}')
```

```python
# 34k blended with 42k
blend_bc42_x1 = x1[-271:-177]
blend_bc42_x2 = x2[-271:-177]
blend_bc42_y1 = y1[-271:-177]
blend_bc42_y2 = y2[-271:-177]

# 42k BCP
bc42_x1 = x1[361:379]
bc42_x2 = x2[361:379]
bc42_y1 = y1[361:379]
bc42_y2 = y2[361:379]

full_bcp42_x1 = np.concatenate([bc34_x1, blend_bc42_x1, 1-bc42_x1])
full_bcp42_x2 = np.concatenate([bc34_x2, blend_bc42_x2, bc42_x2])
full_bcp42_y1 = np.concatenate([bc34_y1, blend_bc42_y1, bc42_y1])
full_bcp42_y2 = np.concatenate([bc34_y2, blend_bc42_y2, bc42_y2])

# Filter for additive ratio > 2
mask = full_bcp42_x2 > 2
# mask = (full_bcp42_x2 > 4) & (full_bcp42_x2 < 5)

filtered_x1_42 = full_bcp42_x1[mask]
filtered_x2_42 = full_bcp42_x2[mask]
filtered_y1_42 = full_bcp42_y1[mask]
filtered_y2_42 = full_bcp42_y2[mask]

# print(f'Length of bc: {len(bc_y2)}, length of hp: {len(hp_y2)}, length of both: {len(combo_x1)}')
```

# Feature importance

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Assume df is your original dataframe
# Define inputs and outputs
input_vars = df.columns[1:-2]   # all columns except last two
output_vars = df.columns[-2:]   # last two columns

# Scale inputs and outputs
scaler_inputs = StandardScaler()
scaler_outputs = StandardScaler()

scaled_inputs = scaler_inputs.fit_transform(df[input_vars])
scaled_outputs = scaler_outputs.fit_transform(df[output_vars])

# Create a scaled dataframe for correlation
df_scaled = pd.DataFrame(
    data = pd.concat([
        pd.DataFrame(scaled_inputs, columns=input_vars),
        pd.DataFrame(scaled_outputs, columns=output_vars)
    ], axis=1)
)

# Compute correlation matrix
cor = df_scaled[input_vars.tolist() + output_vars.tolist()].corr()

# Extract top 10 input features for each output
for output in output_vars:
    print(f"\nTop 10 features correlated with {output}:")
    top_features = cor[output].loc[input_vars].abs().sort_values(ascending=False).head(10)
    for feature, corr_value in top_features.items():
        print(f"{feature}: {corr_value:.3f}")
```
