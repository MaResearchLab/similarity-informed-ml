# Import packages and load data

```python
import pandas as pd
import numpy as np
import time
import sklearn
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from skopt import BayesSearchCV
from skopt.space import Integer, Real, Categorical
from sklearn.model_selection import train_test_split
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler 
from joblib import dump, load
```

# Data pre-processing

```python
# Create a master smiles map containing the 3 maps
smiles_maps = ["poly", "repeat", "monomer"]
similarity_methods = ["morgan", "rdkit_path", "maccs", "rdkit_physchem", "mordred"]
output = 'gisaxs_fwhm'
```

```python
# Create a loop that stores each data set into a dictionary
data_directory = '../../chemical-featurization/likeness_data/'
df_likeness = {}
for name in smiles_maps:
    for method in similarity_methods:
        # Continuous descriptor methods cannot handle the wildcard present in the 'poly' SMILES
        if "poly" in name.lower() and method in ["rdkit_physchem", "mordred"]:
            continue
        # Create temporary copy of each df
        file_path = f"{data_directory}{name}_{method}.xlsx"
        temp_df = pd.read_excel(file_path)

        # Store each df into a dictionary for access later
        df_likeness_key = f"{name}_{method}"
        df_likeness[df_likeness_key] = temp_df
```

# Bayesian-Search

```python
def bayes_train(df, output):

    df_clean = df.dropna()

    inputs = df_clean.iloc[:, 1:-2].copy()
    target = df_clean[f"{output}"]

    X_train, X_test, y_train, y_test = train_test_split(
        inputs, target, test_size=0.2, random_state=10
    )
    
    # Scale the inputs (fit only on training data to avoid data leakage)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    # X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
    X_test_scaled = scaler.transform(X_test)
    # X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)

    RANDOM_SEED = 42
    CV_FOLDS = 5

    search_spaces = {

        # forest size (log scaling explores wide range efficiently)
        "n_estimators": Integer(50, 300),

        # tree complexity
        "max_depth": Integer(3, 20),

        # regularization knobs
        # "min_samples_split": Integer(2, 20),
        # "min_samples_leaf": Integer(1, 12),

        # feature sampling
        "max_features": Categorical(["sqrt", "log2", 1.0, 0.3, 0.5, 0.8]),

        # bootstrap behavior
        "bootstrap": Categorical([True, False]),

        # split criterion
        # "criterion": Categorical(["squared_error", "friedman_mse"]),
    }

    rf = RandomForestRegressor(
        random_state=RANDOM_SEED,
        n_jobs=-1
    )

    opt = BayesSearchCV(
        estimator=rf,
        search_spaces=search_spaces,
        # n_iter=60,                 # number of Bayesian steps
        # scoring="r2",
        cv=CV_FOLDS,
        n_jobs=-1,
        random_state=RANDOM_SEED,
        verbose=0
    )

    # Fit optimizer
    opt.fit(X_train_scaled, y_train)

    best_model = opt.best_estimator_
    best_cv_r2 = opt.best_score_
    best_test_r2 = best_model.score(X_test_scaled, y_test)

    print("CV Train R²:", best_cv_r2)
    print("Test R²:", best_test_r2)
    print("Best Parameters:", opt.best_params_)

    return best_model
```

```python
best_models = {}
for name in smiles_maps:
    for method in similarity_methods:
        # Skip invalid combinations
        if "poly" in name.lower() and method in ["rdkit_physchem", "mordred"]:
            continue

        df_likeness_key = f"{name}_{method}"
        print(f'\nTraining model on: {df_likeness_key}')

        start_time = time.time()
        best_models[df_likeness_key] = bayes_train(df_likeness[df_likeness_key], output)
        end_time = time.time()

        elapsed = (end_time - start_time)/60
        print(f"Time for {df_likeness_key}: {elapsed:.2f} minutes")
```

```python
for name in smiles_maps:
    for method in similarity_methods:
        # Continuous descriptor methods cannot handle the wildcard present in the 'poly' SMILES
        if "poly" in name.lower() and method in ["rdkit_physchem", "mordred"]:
            continue
                
        # Store each df into a dictionary for access later
        df_likeness_key = f"{name}_{method}"        
        dump(best_models[df_likeness_key] , f'bayesian/fwhm/scaled_{output}_{df_likeness_key}.joblib')
```
